```python
from scipy.stats import chisquare
```

```python
def chi2_test(col):
    alpha = 0.05
    stat, p = chisquare(col.value_counts().values)
    if p < alpha:
        return 'H0 Rejetée'
    else :
        return 0
```

```python
from sklearn.preprocessing import LabelEncoder,LabelBinarizer
encoder=LabelEncoder()
def encodage(sampligDf):
    for col in sampligDf.select_dtypes('object').columns:
        if col=="Race":
            encoderB=LabelBinarizer()
            sampligDf[col]=encoderB.fit_transform(sampligDf[col])
        if col=="GenHealth":
            GenHealthMap = {'Very good':3,'Fair':1,'Good':2,'Poor':0,'Excellent':4}
            sampligDf[col]=sampligDf.loc[:,col] = sampligDf[col].map(GenHealthMap)
        if col=="Diabetic":
            DiabeticMap = {'Yes':3 , 'No':0 , 'No, borderline diabetes':1 , 'Yes (during pregnancy)':2 ,}
            sampligDf[col]=sampligDf.loc[:,col] = sampligDf[col].map(DiabeticMap)
        else:
            sampligDf[col]=encoder.fit_transform(sampligDf[col])
    return sampligDf
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
def standardisation(df):
    scaler=StandardScaler()
    X=scaler.fit_transform(df)
    return X
```

```python
def preprocessing(sampligDf):
    sampligDf = encodage(sampligDf)
    X = sampligDf.drop("HeartDisease", axis=1)
    y = sampligDf.loc[:,"HeartDisease"]
    X = standardisation(X)
    print(y.value_counts())
    return X, y
```
